```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

nltk.download('punkt')
nltk.download('stopwords')

# Step 1: Dataset (tweets)
tweets = [
    "I love this product",
    "This is very bad",
    "Amazing experience",
    "Worst service ever",
    "I am very happy",
    "I hate this"
]

labels = [1, 0, 1, 0, 1, 0]  # 1=Positive, 0=Negative

# Step 2: Stopword removal function
stop_words = set(stopwords.words('english'))

def clean_text(text):
    words = word_tokenize(text)
    filtered = [w.lower() for w in words if w.lower() not in stop_words]
    return " ".join(filtered)

cleaned_tweets = [clean_text(t) for t in tweets]

print("Cleaned Tweets:", cleaned_tweets)

# Step 3: Bag of Words
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(cleaned_tweets)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("BoW Matrix:\n", X.toarray())

# Step 4: Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2)

# Step 5: Model Training
model = LogisticRegression()
model.fit(X_train, y_train)

# Step 6: Prediction
test_tweet = ["I not love this service"]
test_clean = [clean_text(test_tweet[0])]
test_vector = vectorizer.transform(test_clean)

prediction = model.predict(test_vector)

print("Prediction:", "Positive" if prediction[0] == 1 else "Negative")
```

```python
while True:
    # Step 1: Take input from user
    user_tweet = input("\nEnter a tweet: ")

    # Step 2: Clean the input
    test_clean = [clean_text(user_tweet)]

    # Step 3: Convert to vector
    test_vector = vectorizer.transform(test_clean)

    # Step 4: Predict sentiment
    prediction = model.predict(test_vector)

    # Step 5: Show result
    if prediction[0] == 1:
        print("Prediction: Positive 😊")
    else:
        print("Prediction: Negative 😡")

    # Step 6: Ask user to continue or exit
    choice = input("\nPress 1 to continue or 2 to exit: ")

    if choice == "2":
        print("Exiting program...")
        break
```
